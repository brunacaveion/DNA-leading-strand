# Find pi value for the 1st line

```python
# get 1st line and organize as a 4x4 matrix considering the order we need, which is every 4 elements is a line of the matrix
line1=df1.iloc[:1]
a1 = np.array([[line1[0], line1[1], line1[2], line1[3]],
             [line1[4], line1[5], line1[6], line1[7]],
             [line1[8], line1[9], line1[10], line1[11]],
             [line1[12], line1[13], line1[14], line1[15]]])
matrix1=np.array(a1).reshape(4,4)
matrix1
```

```python
# transpose the matrix because we need eigenvector from the left
transposta1 = np.transpose(matrix1)
transposta1
```

```python
b1,r1 = np.linalg.eig(transposta1)
b1
```

```python
r1[1]
```

# Generalize the method for the rest of the data

```python
# organizing all lines as the matrix needed and store in one table
lenghtdf1 = list(range(1,9222))
lista=[matrix1]
for i in lenghtdf1:
    line=df1.iloc[i:i+1]
    a = np.array([[line[0], line[1], line[2], line[3]],
             [line[4], line[5], line[6], line[7]],
             [line[8], line[9], line[10], line[11]],
             [line[12], line[13], line[14], line[15]]])
    matrix=np.array(a).reshape(4,4)
    lista.append(matrix)
```

```python
transpostadf1=[]
for n in lista:
    transposta = np.transpose(n)
    transpostadf1.append(transposta)
```

```python
transpostadf1
```

```python
eingenvaluedf1 = []
eingenvectordf1 = []
for matriz in transpostadf1:
    b,r = np.linalg.eig(matriz)
    eingenvaluedf1.append(b)
    eingenvectordf1.append(r)
```

```python
# it looks like the eingenvalue 1 is always the first element in the matrix. We need to validate that
eingenvaluedf1
```

```python
# how to access the first element of the second matrix in eingenvaluedf1
primeiro = eingenvaluedf1[1]
primeiro[0].round()
```

```python
# getting all the first elements of all matrices and putting them on a table
eingenvalues1 = []
for i in lenghtdf1:
    primeiro = eingenvaluedf1[i]
    valor = primeiro[0].round()
    eingenvalues1.append(valor)
```

```python
eingenvalues1
```

```python
# this code shows us all the eingenvalues are 1 for the first element of all the matrices, so we can consider the first 
# eingenvalue as the pi we are looking for
for n in eingenvalues1:
    if n != 1:
        print(n)
```

```python
# table with all the pi functions we where looking for
pidf1 = []
for matriz in eingenvectordf1:
    vetor = matriz[0]
    pidf1.append(vetor)
```

```python
pidf1
```

# is the least DNA a leading strand?

```python
# In order to know if is a leading strand, we need to verify if piG + piT > piA + piC, 
# where pi = [piA piC piG piT] was the eingenvectors found previously
```

```python
test = pidf1[9221]
GTtest = test[2]+test[3]
ACtest = test[0]+test[1]

if GTtest > ACtest:
    print('It is leading strand!')
else:
    print('It is not leading strand')
# Note that GTtest > ACtest, that means the DNA strand on the position 1 of df1 is the leading strand
```

```python
GTtest
```

```python
ACtest
```

# Generalize for all lines and get only the leading strand lines

```python
lenghtdf1 = list(range(1,9222))
classificacaodf1=[]
for i in lenghtdf1:
    matrix = pidf1[i]
    GT = matrix[2]+matrix[3]
    AC = matrix[0]+matrix[1]
    if GT>AC:
        classificacaodf1.append('leading')
    else:
        classificacaodf1.append('no')


    
```

```python
classificador = pd.DataFrame(classificacaodf1)
```

```python
df1['is it leading'] = classificador
```
